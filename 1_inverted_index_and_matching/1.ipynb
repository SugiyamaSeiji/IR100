```python
invert_index = {}

for index, row in product_us.iterrows():
    texts = row['product_title'].split()

    for text in texts:
        if text not in invert_index.keys():
            invert_index[text] = set()
        invert_index[text].add(row['product_id'])
```

### 10. プレフィックス

```python
import os

class PostingListIterator:
    def __init__(self, postings_set):
        self.postings = self.convert_to_postings(postings_set)
        self.compressed_postings = self.compress_postings(self.postings)
        self.current_index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_index < len(self.compressed_postings):
            posting = self.compressed_postings[self.current_index]
            self.current_index += 1
            return self.expand_posting(posting)
        else:
            raise StopIteration

    def convert_to_postings(self, postings_set): #  キーに対するset型の配列をリスト型の配列に変更する(1対1)
        result = []
        for topic, product_ids in postings_set.items():
            for product_id in product_ids:
                result.append((topic, product_id))
        return result

    def compress_postings(self, postings):
        compressed_postings = []
        prev_product_id = ""
        for topic, product_id in postings:
            if prev_product_id:
                common_prefix = os.path.commonprefix([prev_product_id, product_id])
                compressed_postings.append((topic, common_prefix + product_id[len(common_prefix):]))
            else:
                compressed_postings.append((topic, product_id))
            prev_product_id = product_id
        return compressed_postings

    def expand_posting(self, posting):
        topic, product_id = posting
        return topic, product_id
```

```python
iterator = PostingListIterator(invert_index)
```

### 11. 符号化

```python
import os

class SignIterator:
    def __init__(self, postings_set):
        self.postings = self.convert_to_postings(postings_set)
        self.current_index = 0
        self.product_ids = sorted(set(posting[1] for posting in self.postings))
        self.product_id_index = {product_id: i for i, product_id in enumerate(self.product_ids)}
        self.compressed_postings = self.compress_postings(self.postings)

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_index < len(self.compressed_postings):
            posting = self.compressed_postings[self.current_index]
            self.current_index += 1
            return posting
        else:
            raise StopIteration

    def convert_to_postings(self, postings_set):
        result = []
        for topic, product_ids in postings_set.items():
            for product_id in product_ids:
                result.append((topic, product_id))
        return result

    def compress_postings(self, postings):
        compressed_postings = []
        prev_product_id = ""
        for topic, product_id in postings:
            index = self.product_id_index[product_id]
            if prev_product_id:
                common_prefix = os.path.commonprefix([prev_product_id, product_id])
                compressed_postings.append((topic, (index, len(common_prefix), product_id[len(common_prefix):])))
            else:
                compressed_postings.append((topic, (index, 0, product_id)))
            prev_product_id = product_id
        return compressed_postings

    def expand_posting(self, posting):
        topic, (index, prefix_length, suffix) = posting
        product_id = self.product_ids[index][:prefix_length] + suffix
        return topic, product_id

iterator = SignIterator(invert_index)

i = 0
for num in iterator:
    if i == 10:
        break
    i+=1
    print(num)
```

### 12. Variabel Byte Encoding

```python
class PostingListIterator:
    def __init__(self, postings_set):
        self.postings = self.convert_to_postings(postings_set)
        self.current_index = 0
        self.product_ids = sorted(set(posting[1] for posting in self.postings))
        self.product_id_index = {product_id: i for i, product_id in enumerate(self.product_ids)}
        self.compressed_postings = self.compress_postings(self.postings)

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_index < len(self.compressed_postings):
            posting = self.compressed_postings[self.current_index]
            self.current_index += 1
            return posting
        else:
            raise StopIteration

    def convert_to_postings(self, postings_set):
        result = []
        for topic, product_ids in postings_set.items():
            for product_id in product_ids:
                result.append((topic, product_id))
        return result

    def compress_postings(self, postings):
        compressed_postings = []
        prev_index = 0
        for topic, product_id in postings:
            index = self.product_id_index[product_id]
            diff = index - prev_index
            encoded_diff = self.encode_variable_byte(diff)
            compressed_postings.append((topic, encoded_diff))
            prev_index = index
        return compressed_postings

    def encode_variable_byte(self, num):
        bytes_array = []
        while True:
            byte = num % 128
            num //= 128
            if num > 0:
                byte |= 128
            bytes_array.append(byte)
            if num == 0:
                break
        return bytes_array

    def decode_variable_byte(self, bytes_array):
        num = 0
        for i, byte in enumerate(bytes_array):
            num |= (byte & 127) << (7 * i)
            if not byte & 128:
                break
        return num

    def expand_posting(self, posting):
        topic, encoded_diff = posting
        diff = self.decode_variable_byte(encoded_diff)
        index = diff + (prev_index if 'prev_index' in locals() else 0)
        prev_index = index
        product_id = self.product_ids[index]
        return topic, product_id

iterator = PostingListIterator(invert_index)

i = 0
for num in iterator:
    if i == 50:
        break
    i += 1
    print(num)
```
